```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_table("GDS5420.txt").drop("ID_REF",axis = 1).set_index("IDENTIFIER")
dataset = dataset.replace(np.nan,0)
dataset.tail()
```

```python
logged_data = dataset.apply(np.log2)
logged_data = logged_data.replace(-np.inf,0)
logged_data = logged_data.drop(index=["--Control","STAT1"])
logged_data.tail()
```
